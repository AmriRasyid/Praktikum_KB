##### Import Library 

```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("glass_cleaned.csv")
df
```

#### <b> DNN <b>

##### Menentukan Atribut

```python
x_data = df[[
    "RI",
    "Na",
    "Mg",
    "Al",
    "Si",
    "K",
    "Ca",
    "Ba",
    "Fe",
]]

y_data = df["Type"]
```

##### Splitting Data 75% train dan 25% testing

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=.25)
```

```python
num_class = 7
y_train = tf.keras.utils.to_categorical(y_train, num_class)
y_test = tf.keras.utils.to_categorical(y_test, num_class)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(512, activation="relu", input_dim=9),
    tf.keras.layers.Dense(256, activation="selu"),
    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dense(128, activation="elu"),
    tf.keras.layers.Dense(num_class, activation="softmax")
])
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.categorical_crossentropy,
    metrics=["accuracy"]
)
```

```python
model_fit = model.fit(
    x=x_train,
    y=y_train,
    validation_data=(x_test, y_test),
    epochs=200,
    steps_per_epoch=18,
    verbose=2,
)
```

##### Print hasil akurasi

```python
print("Akurasi didapat:", model_fit.history["accuracy"][-1] * 100, "%")
print("Akurasi validasi didapat:", model_fit.history["val_accuracy"][-1] * 100, "%")
```
