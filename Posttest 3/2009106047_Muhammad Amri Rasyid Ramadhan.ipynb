```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("healthcare-dataset-stroke-data.csv")
```

```python
df.head()
```

```python
category = ["ever_married","work_type","Residence_type","smoking_status"]

for i in category:
    print(df[i].value_counts())
    print(" ")
```

## Grafik 1 : Perbandingan Tipe Perkerjaan (Grafik Pie)

```python
dfGroup = df.groupby("work_type")["id"].count()
dfGroup.head()
```

```python
plt.style.use('ggplot')


plt.figure(figsize =(10,10))
plt.pie(dfGroup,labels = dfGroup.index,startangle=90,autopct="%.1f%%",explode=(0,0,0,0,0.05))
plt.title(" Perbandingan Persentase Tipe Perkerjaan")
plt.show()
```

## Grafik 2

```python
plt.figure(figsize =(10,8))

sns.countplot(data = df,x="Residence_type",hue="work_type")
plt.title("Perbandingan jumlah berdasarkan jenis tempat tinggal")
plt.xlabel("Jenis Tempat Tinggal")
plt.ylabel("Jumlah")
plt.show()
```

## Grafik 3

```python
plt.figure(figsize =(10,8))

sns.boxplot(data = df,x="Residence_type",y="bmi")
plt.title("Penyebaran Data BMI Berdasarkan Jenis Tempat Tinggal")
plt.xlabel("Jenis Tempat Tinggal")
plt.ylabel("BMI")
plt.show()
```
