### Data Cleaning

```python
from sklearn.impute import SimpleImputer
```

```python
df.isna().sum()
```

### Mengubah nilai null yang punya tipe data kategorikal

```python
imputer = SimpleImputer(strategy = 'most_frequent')
df["gender"] = imputer.fit_transform(df[["gender"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

### Mengubah nilai null yang punya tipe data bilang bulat

```python
imputer = SimpleImputer(strategy = 'most_frequent')
df["age"] = imputer.fit_transform(df[["age"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

### Mengubah nilai null yang punya tipe data desimal

```python
imputer = SimpleImputer(strategy = 'mean')
df["avg_glucose_level"] = imputer.fit_transform(df[["avg_glucose_level"]])
print("\nJumlah null setelah menggunakan SimpleImputer: ")
print(df.isna().sum())
```

### One hot encoding

```python
from sklearn.preprocessing import OneHotEncoder 
```

```python
encoder = OneHotEncoder(sparse=False) 
maritalEnc = encoder.fit_transform(df[["id"]]) 
maritalEnc = pd.DataFrame(maritalEnc) 
df = df.join(maritalEnc) 
df 
```
